```python
import pandas as pd
import re
```

```python
datashark=pd.read_csv('attacks.csv',encoding="latin",sep=',')
```

```python
datashark.duplicated()
```

```python
datashark_clean=datashark.drop_duplicates()
datashark_clean.isna()
```

```python
datashark_clean=datashark[datashark['Date'].notna()]
```

```python
datashark_clean
```

```python
datashark_clean.info()
```

```python
#validacao de data certa
def pegarmes(x):
    patern=r'[0-9]{1,2}[-][a-zA-Z]{3}[-][0-9]{4}'
    mes=re.findall(patern,str(x))
    return bool(mes)
```

```python
df=datashark_clean[datashark_clean['Date'].apply(pegarmes)==True]
```

```python
#tirar os espacos no inicio e no final do campo country
df['Country']=df['Country'].apply(lambda x:re.sub(r'^ +','',str(re.sub(r' +$','',str(x)))))
df
```

```python
#criando o dataframe de hemisferio
df_hemisphere=pd.read_csv('hemisphere.csv', sep=';')
df_hemisphere['Country']=df_hemisphere['Country'].apply(lambda x: x.upper())
df_hemisphere
```

```python
#escolhendo os campos pra fazer analise
df_shark=df[['Date','Country','Activity','Sex ','Age','Fatal (Y/N)']]
```

```python
#limpando a coluna Date
df_shark['Date']=df_shark['Date'].apply(lambda x: ''.join(re.findall(r'[0-9]{1,2}[-][a-zA-Z]{3}[-][0-9]{4}',str(x))))
```

```python
#unindo os dataframe de shark e hemisferio
df_shark_h=pd.merge(df_shark, df_hemisphere, on="Country", how="inner")
df_shark_h
```

```python
#fazendo a ultima limpeza da coluna date
df_shark_h['Date']=df_shark_h['Date'].apply(lambda x: ''.join(re.findall(r'^[0-9]{1,2}[-][a-zA-Z]{3}[-][0-9]{4}',str(x))))
```

```python
#criando a funcao pra asinar a coluna de estacao do ano
from datetime import date, datetime

def get_season(fecha,h):
    objDate = datetime.strptime(fecha, '%d-%b-%Y').date()
    Y = objDate.year # hemisferio norte
        
    estaciones_n = [('winter', date(Y,  1,  1),  date(Y,  3, 20)),
              ('spring', date(Y,  3, 21),  date(Y,  6, 20)),
              ('summer', date(Y,  6, 21),  date(Y,  9, 22)),
              ('fall', date(Y,  9, 23),  date(Y, 12, 20)),
              ('winter', date(Y, 12, 21),  date(Y, 12, 31))]
    estaciones_s = [('summer', date(Y,  1,  1),  date(Y,  3, 20)),
              ('fall', date(Y,  3, 21),  date(Y,  6, 20)),
              ('winter', date(Y,  6, 21),  date(Y,  9, 22)),
              ('spring', date(Y,  9, 23),  date(Y, 12, 20)),
              ('summer', date(Y, 12, 21),  date(Y, 12, 31))]
    
    if h=='northern':
        for estacion, inicio, fin in estaciones_n:
            if inicio <= objDate <= fin:
                return estacion
    elif h=='southern':
        for estacion, inicio, fin in estaciones_s:
            if inicio <= objDate <= fin:
                return estacion
    else:
        estacion='summer'
        return estacion
```

```python
#criando a coluna season
df_shark_h['Season']=df_shark_h[["Date", "Hemisphere"]].apply(lambda x: get_season(*x), axis=1)
```

```python
df_shark_h
```

```python
list_top_c=list(df_shark_h['Country'].value_counts().head(5).index)
df_shark_c=df_shark_h
df_shark_c['Country']=df_shark_h['Country'].apply(lambda x: 'Rest' if x not in list_top_c else x)
```

```python
df_shark_c
```

```python
df_shark_c.groupby(['Country']).count().plot(
    kind='pie', y='Date', autopct='%1.0f%%',startangle=60,figsize=(15,10),title='SHARK ATTACK REPORT',fontsize=20)
```

```python
df_shark_h.groupby(['Season']).count().plot(
    kind='pie', y='Country', autopct='%1.0f%%',startangle=60,figsize=(15,10),title='WORLD',fontsize=20)
```

```python
#limpeza da coluna Fatal
df_shark_h['Fatal (Y/N)']=df_shark_h['Fatal (Y/N)'].apply(lambda x: 'N' if x!='Y' else 'Y')
df_shark_h['Fatal (Y/N)'].value_counts()
```

```python
df_summer=df_shark_h[df_shark_h['Season']=='summer']
df_summer.groupby(['Fatal (Y/N)']).count().plot(
    kind='pie', y='Country', autopct='%1.0f%%',startangle=60,figsize=(15,10),title='Fatal (Y/N) SUMMER',fontsize=20)
df_summer['Fatal (Y/N)'].value_counts()
```

```python
df_winter=df_shark_h[df_shark_h['Season']=='winter']
df_winter.groupby(['Fatal (Y/N)']).count().plot(
    kind='pie', y='Country', autopct='%1.0f%%',startangle=60,figsize=(15,10),title='Fatal (Y/N) WINTER',fontsize=20)
df_winter['Fatal (Y/N)'].value_counts()
```

```python
df_spring=df_shark_h[df_shark_h['Season']=='spring']
df_spring.groupby(['Fatal (Y/N)']).count().plot(
    kind='pie', y='Country', autopct='%1.0f%%',startangle=60,figsize=(15,10),title='Fatal (Y/N) SPRING',fontsize=20)
df_spring['Fatal (Y/N)'].value_counts()
```

```python
df_fall=df_shark_h[df_shark_h['Season']=='fall']
df_fall.groupby(['Fatal (Y/N)']).count().plot(
    kind='pie', y='Country', autopct='%1.0f%%',startangle=60,figsize=(15,10),title='Fatal (Y/N) FALL',fontsize=20)
df_fall['Fatal (Y/N)'].value_counts()
```
