```python
from datetime import date, datetime

def get_season(fecha,h):
    objDate = datetime.strptime(fecha, '%d-%b-%Y').date()
    Y = objDate.year # hemisferio norte
        
    estaciones_n = [('winter', date(Y,  1,  1),  date(Y,  3, 20)),
              ('spring', date(Y,  3, 21),  date(Y,  6, 20)),
              ('summer', date(Y,  6, 21),  date(Y,  9, 22)),
              ('fall', date(Y,  9, 23),  date(Y, 12, 20)),
              ('winter', date(Y, 12, 21),  date(Y, 12, 31))]
    estaciones_s = [('summer', date(Y,  1,  1),  date(Y,  3, 20)),
              ('fall', date(Y,  3, 21),  date(Y,  6, 20)),
              ('winter', date(Y,  6, 21),  date(Y,  9, 22)),
              ('spring', date(Y,  9, 23),  date(Y, 12, 20)),
              ('summer', date(Y, 12, 21),  date(Y, 12, 31))]
    
    if h=='northern':
        for estacion, inicio, fin in estaciones_n:
            if inicio <= objDate <= fin:
                return estacion
    elif h=='southern':
        for estacion, inicio, fin in estaciones_s:
            if inicio <= objDate <= fin:
                return estacion
    else:
        estacion='summer'
        return estacion
    
```
